```python
#o tambien podemos crear graficas tan impresionantes como esta

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
X = np.arange(-5, 5, 0.15)
Y = np.arange(-5, 5, 0.15)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis')
plt.show()
```

# UNIDAD 9.- OTRAS GRAFICAS


La biblioteca Matplotlib en Python puede crear gráficos de burbujas, que son una variante de los gráficos de dispersión. En un gráfico de burbujas, los datos se presentan como una colección de puntos en el plano, donde la posición de cada punto está determinada por las variables indicadas en los ejes X e Y, y el tamaño de cada punto (o burbuja) puede representar una tercera variable. Esto permite visualizar tres dimensiones de datos en un gráfico bidimensional.

Tambien puede mostrar una cuarta variable, este código, colors representa los valores de la cuarta variable, y el argumento cmap='viridis' especifica el mapa de colores utilizado para visualizar esta variable. La función plt.colorbar() agrega una barra de colores al gráfico, lo que ayuda a interpretar el significado de los colores de las burbujas.

Con esta técnica, puedes visualizar hasta cuatro dimensiones de datos: las posiciones X e Y, el tamaño de las burbujas y el color de las burbujas, lo que permite un análisis más profundo y una representación más rica de tu conjunto de datos.

MAPA DE CALOR

```python
# Mapa de calor
# Para crear un mapa de calor en Matplotlib, podemos utilizar la función imshow()
# para visualizar los datos en un arreglo 2D, donde los colores representan los
# valores de los datos. A continuación, te muestro cómo crear un mapa de calor
# utilizando datos sintéticos:

import matplotlib.pyplot as plt
import numpy as np

# Generar datos sintéticos
data = np.random.rand(10, 10) # Matriz de 10x10 con valores aleatorios

# El parámetro interpolation controla cómo se interpolarán los píxeles de la
# imagen cuando se muestre. En este caso, se utiliza 'nearest', lo que significa
# que cada píxel de la imagen se asignará al píxel más cercano en la pantalla.
# Esto se utiliza para evitar suavizar demasiado la imagen y conservar la
# claridad en la representación de los datos.
# cmap es el mapa de color
plt.imshow(data, cmap='Blues', interpolation='nearest')
plt.xlabel('Venta de productos')
plt.ylabel('Ubicación de la Tienda')
plt.title('Mapa de Calor de las ventas')
plt.colorbar() # Añadir barra de color para indicar la escala de valores
plt.show()
```

```python
# Supongamos que tienes 4 animales: un gato, un perro, un tigre y un león.

# Similitudes: Primero, notas que el gato y el perro son bastante diferentes
# entre sí, pero menos diferentes entre ellos que de un tigre o un león. Así que
# los pones en grupos separados pero cercanos. Luego, ves que el tigre y el león
# son muy similares, así que los agrupas juntos.

# Agrupamiento: Ahora, tienes dos grupos: {gato, perro} y {tigre, león}. Aunque
# estos grupos son diferentes, sabes que todos son mamíferos y tienen cuatro
# patas, así que tienen algunas cosas en común.

# Dendrograma: Para visualizar esto, dibujas un diagrama que comienza con líneas
# que conectan al tigre y al león, ya que son los más similares. Luego, esas
# líneas se conectan a una línea más alta que también se conecta a las líneas
# del gato y del perro, mostrando que todos estos animales están relacionados,
# pero a diferentes niveles de similitud.

from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
import numpy as np

# Generar datos sintéticos: 10 puntos en un espacio de 2 dimensiones
np.random.seed(42) # Para reproducibilidad
data = np.random.multivariate_normal([0, 0], [[1, 0], [0, 1]], 10)

# Realizar agrupamiento jerárquico
Z = linkage(data, 'ward') # 'ward' es uno de los métodos de agrupamiento

# Generar y mostrar el dendrograma
plt.figure(figsize=(10, 7))
plt.title('Dendrograma')
plt.xlabel('Índice de la Muestra')
plt.ylabel('Distancia de Ward')
dendrogram(Z)
plt.show()
```

```python
#Arboles de decision
# Un árbol de decisión es como un mapa que te ayuda a tomar decisiones basándose
# en una serie de preguntas y respuestas. Imagina que estás tratando de decidir
# qué actividad hacer en un día libre. El árbol comienza con una pregunta inicial,
# como "¿Está lloviendo?". Si la respuesta es sí, el árbol te guía hacia
# actividades en interiores; si la respuesta es no, te sugiere actividades al
# aire libre. A partir de ahí, puedes seguir haciendo más preguntas como
# "¿Tengo equipo deportivo?" o "¿Prefiero leer?", y el árbol te guiará hasta
# llegar a una actividad sugerida basada en tus respuestas.

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Cargar el conjunto de datos
iris = load_iris()

# Entrenar el modelo de árbol de decisión
clf = DecisionTreeClassifier(random_state=123)
clf = clf.fit(iris.data, iris.target)

# Crear la gráfica del árbol de decisión usando Matplotlib
plt.figure(figsize=(20,10))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names, rounded=True)
plt.show()
```

GRAFICA 3D

```python
import matplotlib.pyplot as plt
import numpy as np

# Generamos datos aleatorios para x, y, z
mean = 3
number = 50  #numero de puntos que generamos
#creamos 3 arrays o arreglos con  media de 3 y desviacion de 1
x1 = np.random.normal(mean, 1, size=number)
y1 = np.random.normal(mean, 1, size=number)
z1 = np.random.normal(mean, 1, size=number)

# Creamos una figura y ejes 3D
fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(111, projection="3d")

# Añade un conjunto de ejes a la figura fig con una configuración específica.
# 111 es un código que significa que la figura debe contener solo un subplot
# (es decir, una grilla 1x1) y que este subplot será el primero (y único) en la
# figura. projection="3d" indica que el conjunto de ejes será utilizado para
# una gráfica tridimensional. ax es el objeto de los ejes que se utilizará para
# dibujar la gráfica 3D.
#

# Graficamos los puntos de dispersión
# ax: representa el espacio tridimensional donde se va a dibujar la gráfica.
# scatter: Es el método utilizado para crear una gráfica de dispersión.
# (x1, y1, z1): Estos son los arreglos o listas de valores que representan las
# coordenadas de los puntos en el espacio 3D.
# s=100: Este parámetro especifica el tamaño de los puntos en la gráfica de
# dispersión.
ax.scatter(x1, y1, z1, s=100)

# Etiquetamos los ejes
#debes usar set_xlabel, set_ylabel, y set_zlabel para etiquetar los ejes
#x, y, z en gráficas 3D con matplotlib
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

# Mostramos el gráfico
plt.show()
```
