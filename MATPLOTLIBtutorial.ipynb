```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Podemos crear graficas tan sencillas como esta
x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

plt.plot(x, y)

plt.show()
```

```python
#o tambien podemos crear graficas tan impresionantes como esta

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
X = np.arange(-5, 5, 0.15)
Y = np.arange(-5, 5, 0.15)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis')
plt.show()
```

# UNIDAD 9.- OTRAS GRAFICAS


La biblioteca Matplotlib en Python puede crear grficos de burbujas, que son una variante de los grficos de dispersin. En un grfico de burbujas, los datos se presentan como una coleccin de puntos en el plano, donde la posicin de cada punto est determinada por las variables indicadas en los ejes X e Y, y el tamao de cada punto (o burbuja) puede representar una tercera variable. Esto permite visualizar tres dimensiones de datos en un grfico bidimensional.

Tambien puede mostrar una cuarta variable, este cdigo, colors representa los valores de la cuarta variable, y el argumento cmap='viridis' especifica el mapa de colores utilizado para visualizar esta variable. La funcin plt.colorbar() agrega una barra de colores al grfico, lo que ayuda a interpretar el significado de los colores de las burbujas.

Con esta tcnica, puedes visualizar hasta cuatro dimensiones de datos: las posiciones X e Y, el tamao de las burbujas y el color de las burbujas, lo que permite un anlisis ms profundo y una representacin ms rica de tu conjunto de datos.

```python
# Mapa de calor
# Para crear un mapa de calor en Matplotlib, podemos utilizar la funcin imshow()
# para visualizar los datos en un arreglo 2D, donde los colores representan los
# valores de los datos. A continuacin, te muestro cmo crear un mapa de calor
# utilizando datos sintticos:

import matplotlib.pyplot as plt
import numpy as np

# Generar datos sintticos
data = np.random.rand(10, 10) # Matriz de 10x10 con valores aleatorios

# El parmetro interpolation controla cmo se interpolarn los pxeles de la
# imagen cuando se muestre. En este caso, se utiliza 'nearest', lo que significa
# que cada pxel de la imagen se asignar al pxel ms cercano en la pantalla.
# Esto se utiliza para evitar suavizar demasiado la imagen y conservar la
# claridad en la representacin de los datos.
# cmap es el mapa de color
plt.imshow(data, cmap='Blues', interpolation='nearest')
plt.xlabel('Venta de productos')
plt.ylabel('Ubicacin de la Tienda')
plt.title('Mapa de Calor de las ventas')
plt.colorbar() # Aadir barra de color para indicar la escala de valores
plt.show()
```

```python
# Supongamos que tienes 4 animales: un gato, un perro, un tigre y un len.

# Similitudes: Primero, notas que el gato y el perro son bastante diferentes
# entre s, pero menos diferentes entre ellos que de un tigre o un len. As que
# los pones en grupos separados pero cercanos. Luego, ves que el tigre y el len
# son muy similares, as que los agrupas juntos.

# Agrupamiento: Ahora, tienes dos grupos: {gato, perro} y {tigre, len}. Aunque
# estos grupos son diferentes, sabes que todos son mamferos y tienen cuatro
# patas, as que tienen algunas cosas en comn.

# Dendrograma: Para visualizar esto, dibujas un diagrama que comienza con lneas
# que conectan al tigre y al len, ya que son los ms similares. Luego, esas
# lneas se conectan a una lnea ms alta que tambin se conecta a las lneas
# del gato y del perro, mostrando que todos estos animales estn relacionados,
# pero a diferentes niveles de similitud.

from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
import numpy as np

# Generar datos sintticos: 10 puntos en un espacio de 2 dimensiones
np.random.seed(42) # Para reproducibilidad
data = np.random.multivariate_normal([0, 0], [[1, 0], [0, 1]], 10)

# Realizar agrupamiento jerrquico
Z = linkage(data, 'ward') # 'ward' es uno de los mtodos de agrupamiento

# Generar y mostrar el dendrograma
plt.figure(figsize=(10, 7))
plt.title('Dendrograma')
plt.xlabel('ndice de la Muestra')
plt.ylabel('Distancia de Ward')
dendrogram(Z)
plt.show()
```

```python
#Arboles de decision
# Un rbol de decisin es como un mapa que te ayuda a tomar decisiones basndose
# en una serie de preguntas y respuestas. Imagina que ests tratando de decidir
# qu actividad hacer en un da libre. El rbol comienza con una pregunta inicial,
# como "Est lloviendo?". Si la respuesta es s, el rbol te gua hacia
# actividades en interiores; si la respuesta es no, te sugiere actividades al
# aire libre. A partir de ah, puedes seguir haciendo ms preguntas como
# "Tengo equipo deportivo?" o "Prefiero leer?", y el rbol te guiar hasta
# llegar a una actividad sugerida basada en tus respuestas.

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Cargar el conjunto de datos
iris = load_iris()

# Entrenar el modelo de rbol de decisin
clf = DecisionTreeClassifier(random_state=123)
clf = clf.fit(iris.data, iris.target)

# Crear la grfica del rbol de decisin usando Matplotlib
plt.figure(figsize=(20,10))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names, rounded=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generamos datos aleatorios para x, y, z
mean = 3
number = 50  #numero de puntos que generamos
#creamos 3 arrays o arreglos con  media de 3 y desviacion de 1
x1 = np.random.normal(mean, 1, size=number)
y1 = np.random.normal(mean, 1, size=number)
z1 = np.random.normal(mean, 1, size=number)

# Creamos una figura y ejes 3D
fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(111, projection="3d")

# Aade un conjunto de ejes a la figura fig con una configuracin especfica.
# 111 es un cdigo que significa que la figura debe contener solo un subplot
# (es decir, una grilla 1x1) y que este subplot ser el primero (y nico) en la
# figura. projection="3d" indica que el conjunto de ejes ser utilizado para
# una grfica tridimensional. ax es el objeto de los ejes que se utilizar para
# dibujar la grfica 3D.
#

# Graficamos los puntos de dispersin
# ax: representa el espacio tridimensional donde se va a dibujar la grfica.
# scatter: Es el mtodo utilizado para crear una grfica de dispersin.
# (x1, y1, z1): Estos son los arreglos o listas de valores que representan las
# coordenadas de los puntos en el espacio 3D.
# s=100: Este parmetro especifica el tamao de los puntos en la grfica de
# dispersin.
ax.scatter(x1, y1, z1, s=100)

# Etiquetamos los ejes
#debes usar set_xlabel, set_ylabel, y set_zlabel para etiquetar los ejes
#x, y, z en grficas 3D con matplotlib
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

# Mostramos el grfico
plt.show()
```
